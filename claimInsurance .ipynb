```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("insurance2.csv")
df.sample(4)
```

```python
df.drop(['region'],axis=1,inplace=True)
df
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(df[['age','sex','bmi','children','smoker','charges']],df.insuranceclaim,test_size=0.2,random_state=25)
```

```python
X_test.max()
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train_scaled['age']/100
X_train_scaled['bmi'] = X_train_scaled['bmi']/49
X_train_scaled['children'] = X_train_scaled['children']/5
X_train_scaled['charges'] = X_train_scaled['charges'] / 48517.56315


X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test_scaled['age'] /100
X_test_scaled['bmi'] = X_test_scaled['bmi']/49
X_test_scaled['children'] = X_test_scaled['children']/5
X_test_scaled['charges'] = X_test_scaled['charges'] / 48517.56315
```

```python
X_train_scaled
```

```python
model = keras.Sequential([
    keras.layers.Dense(200,input_shape=(6,),activation='relu',kernel_initializer='ones',bias_initializer='zeros'),
    keras.layers.Dense(100,activation='tanh'),
    keras.layers.Dense(1,activation='sigmoid')  
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
              metrics=['accuracy']
              )
model.fit(X_train_scaled,y_train,epochs=1000)
```

```python
model.evaluate(X_test_scaled,y_test)
```
